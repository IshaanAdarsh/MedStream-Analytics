```python
import matplotlib.pyplot as plt
import random
import time
import seaborn as sns
from IPython.display import display, clear_output

# Initialize empty lists to store data
time_values = []
heart_rate_values = []
temperature_values = []
blood_pressure_values = []

# Create a function to update the real-time plot
def update_plot():
    plt.figure(figsize=(12, 6))
    
    plt.subplot(311)
    sns.lineplot(x=time_values, y=heart_rate_values)
    plt.xlabel('Time')
    plt.ylabel('Heart Rate (BPM)')
    
    plt.subplot(312)
    sns.lineplot(x=time_values, y=temperature_values)
    plt.xlabel('Time')
    plt.ylabel('Temperature (Celsius)')
    
    plt.subplot(313)
    sns.lineplot(x=time_values, y=blood_pressure_values)
    plt.xlabel('Time')
    plt.ylabel('Systolic Blood Pressure (mmHg)')
    
    plt.tight_layout()
    clear_output(wait=True)
    display(plt.gcf())

# Set Seaborn style
sns.set_style("darkgrid")

try:
    while True:
        timestamp = time.strftime("%H:%M:%S")
        
        # Generate random measurements
        heart_rate = random.randint(60, 100)
        temperature = random.uniform(36.5, 37.5)
        blood_pressure = random.randint(100, 160)
        
        # Append data to lists
        time_values.append(timestamp)
        heart_rate_values.append(heart_rate)
        temperature_values.append(temperature)
        blood_pressure_values.append(blood_pressure)
        
        # Keep a limited number of data points for visualization
        if len(time_values) > 50:
            time_values.pop(0)
            heart_rate_values.pop(0)
            temperature_values.pop(0)
            blood_pressure_values.pop(0)
        
        update_plot()
        time.sleep(1)  # Update the plot every second

except KeyboardInterrupt:
    pass

```
